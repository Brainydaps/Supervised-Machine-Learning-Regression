# What is machine learning?


You might recall your secondary school days learning math, particularly those abstract concepts like matrices, calculus, and other topics that seemed disconnected from everyday life. You probably wondered why you needed to endure the challenge of understanding these topics, suspecting you might never use them in real life.

Well, now is the time to see their relevance. Those very mathematical concepts you might have dreaded are the foundation of modern technology and machine learning.

Consider the sophisticated functionalities of image detection on your phone’s camera, text prediction on your keyboard, image generation on Meta AI in WhatsApp, movie recommendations on Netflix, and music suggestions on Spotify. All these impressive capabilities are powered by the application of mathematical principles programmed into computers to enable them to learn.

---

Now, you might be wondering, how do abstract mathematical concepts like algebra, matrices, and calculus contribute to making a computer intelligent or capable of learning?

Before we can make a computer intelligent, we must define intelligence in mathematical terms.

The first challenge was answering the question:
> "What is intelligence?"

Most people would agree that:

> *Intelligence is the ability to learn.*

This straightforward definition provides a foundational basis for creating intelligent machines.

The next challenge was defining:

> *What is learning?*




How can we determine when learning has occurred?

Fortunately, we have mathematical tools to measure learning, which helps us ascertain if learning has indeed taken place.

This is where the *loss function* becomes crucial.

The loss function is a mathematical formula that allows a computer to evaluate if its current guess (or prediction) is better than previous ones, by comparing and recording past errors to continuously improve.

A common loss function used in regression is the Mean Squared Error (MSE), which is given by:

 $$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

And here’s the breakdown of each element:

$$( n )$$: represents the number of data points.
$$(y_i)$$: reprisents the actual value for the (i)th data point.
$$(\hat{y}_i)$$: reprisents the predicted value for the (i)th data point.
$$(\sum)$$: The summation symbol, indicating that we sum the squared differences for all data points.
$$((y_i - \hat{y}_i)^2)$$: represents the squared difference between the actual and predicted values for the (i)th data point.

Indeed, intelligent computers utilize trial and error, using the loss function to judge whether their guesses are improving or worsening over time. And as we all know, a computer is extremely fast and accurate in making millions and billions of calculations, so trial and error method would go faster than humanly possible.

The thing is when it comes to speed and accuracy, computers reign supreme over humans. But when it comes to actual smartness, humans reigns supreme over a machine. The missing piece of smartness is the mathematical concepts that have been coded into a computer to learn and become "smart"


The final step is determining when learning has been maximized and is sufficient.

To address this, a computer is programmed to cease its trial and error after a certain number of attempts when the loss function no longer shows significant improvement.

---

By understanding these principles, you can appreciate how those abstract mathematical concepts you learned in school are crucial in making machines intelligent and capable of learning.


## What Does a Machine Learn From?

Now the next logical question to ask is:

*What does a machine learn from?*

We all know that a machine (computer) does not sense the world the way living people do.

A machine does not smell, hear, taste, see, or feel the way a living person does. So what does a machine learn from if it doesn't sense the world the way humans do?

This is where datasets come into play.

A *dataset* is a structured file that contains data (information), appropriately labeled for a machine to understand and manipulate.

*Data* is how a computer senses and understands the world.

When you take a picture with your smartphone, your camera sensors capture light from the outside world and convert it into multiple pixels that are placed next to each other in the form of a picture format (JPEG, TIFF, PNG, etc.). Each pixel is a string of characters, and each character is a string of binary that the computer actually "senses" and understands. The raw form of that picture will just be gibberish to a human eye.

The same thing applies when you record your voice with your smartphone. Your microphone sensors detect the unique vibrations in the air and convert those vibrations into audio data, which is then saved in an audio format file like MP3 that can be played by another computer device. The basic unit of audio data is a sample, and these samples record the amplitude of the sound wave at specific intervals. A common sample rate is 44,100 samples per second, which is equivalent to 44 kHz. These amplitudes are a series of numbers that are converted to binaries which the computer actually understands.

So everything you do on your smartphones/computers is all data to a computer. Every file, every piece of media, even websites, are all pieces of data that a computer senses in binary.

---

You might notice something: everything you do on your computer/smartphones is eventually converted to numbers for it to pass through a computer. Yes, you heard that right. Even the video you watch on YouTube is just a series of numbers to a computer. A computer does not understand anything other than numbers.

---

You might then be wondering, what about datasets that contain pictures, alphabetical text, even videos? How does the computer manipulate these things that are "obviously not numbers"?

This is where featurization comes into play.

*Featurization* is the conversion of all data types into numbers (integers) that a computer can use. This conversion is done by your computer in your data pipelines before proceeding to process, manipulate, and learn from your dataset containing things that are not numbers.

## How Does Featurization Work?

Now that we've established that computers need data in numerical form to process and learn, let's dive into the process of *featurization*.

*Featurization* is the process of transforming raw data into numerical features that can be used by machine learning algorithms. This step is crucial because it allows computers to interpret, analyze, and learn from various types of data, including text, images, audio, and more.

### Steps in Featurization:

1. *Data Collection*:
    - The first step is gathering raw data. This data can come from various sources such as sensors, databases, or the internet. It can be in different formats, including text files, images, audio recordings, or video.

2. *Data Cleaning*:
    - Before transforming the data, it must be cleaned. This involves removing any noise or irrelevant information, handling missing values, and correcting any inconsistencies.

3. *Data Transformation*:
    - *Text Data*: Text data is often transformed using techniques like tokenization, which breaks down the text into individual words or tokens. These tokens are then converted into numerical values using methods like:
        - *Bag of Words (BoW)*: This approach counts the occurrences of each word in a text.
        - *Term Frequency-Inverse Document Frequency (TF-IDF)*: This technique measures the importance of a word in a document relative to a collection of documents.
        - *Word Embeddings*: Advanced methods like Word2Vec or GloVe represent words in a continuous vector space.
            - *Continuous Vector Space*: This means that words are represented as vectors (arrays of numbers) in a multi-dimensional space, where similar words have vectors that are close together. For example, the words "king" and "queen" might have vectors that are similar but differ in specific dimensions that capture gender. This allows the model to understand and capture the semantic relationships between words.
    - *Image Data*: Images are converted into numerical matrices where each pixel value represents the intensity of light. Techniques like resizing, normalization, and convolutional neural networks (CNNs) can be applied to extract features.
    - *Audio Data*: Audio files are broken down into samples. Techniques such as Fourier transforms can convert these samples into frequency domain features, capturing patterns in the sound waves.
    - *Categorical Data*: Categorical data (e.g., colors, types) is transformed using methods like:
        - *One-Hot Encoding*: Converts categories into binary vectors.
            - *Binary Vectors*: In one-hot encoding, each category is represented by a vector that has a length equal to the number of categories. Each vector has all elements set to 0 except for one element that is set to 1, indicating the presence of that specific category. For example, if we have three categories ["red", "green", "blue"], "red" could be represented as [1, 0, 0], "green" as [0, 1, 0], and "blue" as [0, 0, 1].
        - *Label Encoding* : Assigns each category a unique integer.


4. *Feature Scaling*:
    - Once the data is transformed into numerical form, it is often scaled to ensure that all features contribute equally to the analysis. Common scaling techniques include:
        - *Normalization*: Rescales the data to a range of [0, 1].
        - *Standardization*: Centers the data around the mean with a standard deviation of 1.

5. *Feature Selection*:
    - To improve model performance and reduce complexity, it's essential to select the most relevant features. Techniques like Principal Component Analysis (PCA), Recursive Feature Elimination (RFE), and correlation analysis can help identify the most impactful features.

### Data Cleaning



Once the raw data is collected, the next crucial step is data cleaning. This process involves preparing the data by removing or correcting any inaccuracies, inconsistencies, and irrelevant parts to ensure that the dataset is of high quality. Clean data is essential for building reliable and accurate machine learning models.

- *Steps in Data Cleaning*:
    - *Handling Missing Values*:
        - Identify missing values in the dataset.
        - Decide how to handle them:
            - *Removal*: Remove rows or columns with missing values if they are not significant.
            - *Imputation*: Fill in missing values using techniques like mean, median, mode, or using more sophisticated methods such as k-nearest neighbors (KNN) or regression.

    - *Removing Duplicates*:
        - Identify and remove duplicate rows to ensure each data point is unique.

    - *Correcting Inconsistencies*:
        - Standardize the format of data entries (e.g., date formats, capitalization, unit measures).
        - Correct any obvious errors (e.g., typos, mislabeling).

    - *Filtering Outliers*:
        - Identify outliers that deviate significantly from other data points.
        - Decide whether to remove or transform outliers based on their impact on the model.

    - *Handling Irrelevant Data*:
        - Remove any data that does not contribute to the analysis or the problem at hand.
        
    - *Normalization and Standardization*:
        - Normalize or standardize numerical features to bring them onto a similar scale, which helps improve the performance of some machine learning algorithms.

    - *Encoding Categorical Data*:
        - Convert categorical data into numerical values using techniques such as one-hot encoding or label encoding to make them suitable for machine learning algorithms.

### Data Transformation

Here’s a detailed explanation of different transformation techniques with examples:

### Text Data

Text data is often transformed using techniques like tokenization, which breaks down the text into individual words or tokens. These tokens are then converted into numerical values using methods like:

#### Bag of Words (BoW)

This approach counts the occurrences of each word in a text.

**Example:**






````python
from sklearn.feature_extraction.text import CountVectorizer

texts = ["I love machine learning", "Machine learning is fun"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)
print(X.toarray())
print(vectorizer.get_feature_names_out())
````

Term Frequency-Inverse Document Frequency (TF-IDF)
This technique measures the importance of a word in a document relative to a collection of documents.

Example:

````python
from sklearn.feature_extraction.text import TfidfVectorizer

texts = ["I love machine learning", "Machine learning is fun"]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
print(X.toarray())
print(vectorizer.get_feature_names_out())
````

Word Embeddings
Advanced methods like Word2Vec or GloVe represent words in a continuous vector space.

Example:

````python
from gensim.models import Word2Vec

sentences = [["I", "love", "machine", "learning"], ["Machine", "learning", "is", "fun"]]
model = Word2Vec(sentences, vector_size=10, window=2, min_count=1, workers=4)
print(model.wv['machine'])
print(model.wv['Machine'])
````

Continuous Vector Space
Words are represented as vectors (arrays of numbers) in a multi-dimensional space, where similar words have vectors that are close together. For example, the words "king" and "queen" might have vectors that are similar but differ in specific dimensions that capture gender.

Example:

````python
# Assuming Word2Vec model from above
machine_vector = model.wv['machine']
learning_vector = model.wv['learning']
print(machine_vector)
print(learning_vector)
````

### Feature Scaling

Once the data is transformed into numerical form, it is often scaled to ensure that all features contribute equally to the analysis. Feature scaling is crucial because many machine learning algorithms perform better when numerical features are on a similar scale. Common scaling techniques include normalization and standardization.

#### Normalization

Normalization rescales the data to a range of [0, 1]. This technique is useful when the data needs to be bounded within a specific range. It assumes the data follows a normal distribution.

**Example:**


````python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Sample data
data = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the data
normalized_data = scaler.fit_transform(data)
print(normalized_data)
````

In this example, each feature value is scaled to the range [0, 1].

Standardization:

Standardization centers the data around the mean with a standard deviation of 1. This is a more bouyant method and can handle any type of data distribution.

Example:

````python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Sample data
data = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the data
standardized_data = scaler.fit_transform(data)
print(standardized_data)
````

In this example, each feature value is transformed to have a mean of 0 and a standard deviation of 1.

Why Feature Scaling is Important
Improves Model Performance: Many machine learning algorithms, such as gradient descent-based methods, k-nearest neighbors, and support vector machines, perform better when features are on a similar scale. (gradient descent, k nearest neighbors and support vector machines are all explained at the end of this subsection)

Speeds Up Convergence: Scaling can help speed up the convergence of gradient descent by ensuring that all features are treated equally. (convergence of gradient descent is explained at the end of this subsection)

Avoids Dominance: Without scaling, features with larger magnitudes can dominate the learning process, leading to biased results.

Choosing Between Normalization and Standardization

Normalization: Use when you need to bound features within a specific range. Suitable for algorithms like k-nearest neighbors and neural networks that are sensitive to the scale of input data.

Standardization: Use when the data follows a Gaussian distribution. Suitable for algorithms like linear regression, logistic regression, and principal component analysis (PCA).



It is important to know that not all models requires scaling to optimize their performance.

## Models Critically Affected by Feature Scaling

### Support Vector Machines (SVM)
The SVM model is sensitive to the relative scales of features because it relies on the distances between data points.

### K-Nearest Neighbors (KNN)
KNN uses distance-based calculations to identify the nearest neighbors, so unscaled features can lead to biased distances.

### Principal Component Analysis (PCA)
PCA is a dimensionality reduction technique that depends on the variance of features, making it sensitive to scaling.

### Neural Networks
Neural networks converge faster and more efficiently when features are on a similar scale because they use gradient descent optimization.

### K-Means Clustering
K-Means clustering is based on distance calculations, so unscaled features can result in distorted clusters.

### Linear Regression (with regularization)
While linear regression itself may not always require scaling, it becomes necessary when regularization (e.g., Ridge or Lasso regression) is applied.

### Gradient Boosting Machines (GBM)
Gradient boosting models can benefit from feature scaling as it helps with convergence during training.

## Models Not Critically Affected by Feature Scaling

### Decision Trees
Decision trees split data based on feature values and are not sensitive to the scale of features.

### Random Forests
As an ensemble of decision trees, random forests inherit the insensitivity to feature scaling.

### Naive Bayes
Naive Bayes classifiers rely on the probability of features rather than their scale.

### Logistic Regression (without regularization)
Logistic regression is less affected by scaling unless regularization is involved, which can introduce sensitivity to the scale of features.

- **Reduced Standard Deviation of Gradients:** Scaling reduces the standard deviation of the gradients of the features, bringing all features' slopes closer together. This makes the optimization process more stable and efficient.

# What exactly is regression?


In mathematical terms, regression is a statistical method used to model and analyze the relationship between a dependent variable(label y) and one or more independent variables(features x1,x2,x3.....). The goal is to estimate the relationship so that we can predict the dependent variable based on the values of the independent variables.

Here are the general formulas for some of the major types of regressions:

1. **Linear Regression**:

   $$
   y = \beta_0 + \beta_1 x + \epsilon
   $$

   where \( $y$ \) is the dependent variable, \( $x$ \) is the independent variable, \( $ \beta_0 $ \) is the intercept, \( $ \beta_1 $ \) is the slope, and \( $\epsilon $ \) is the error term.


2. **Multiple Linear Regression**:
   $$
   y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n + \epsilon
   $$
   where \( $y$ \) is the dependent variable, \( $x_1, x_2, \ldots, x_n $\) are the independent variables, \( $\beta_0 $\) is the intercept, \( $\beta_1 $, $\beta_2 $, $\ldots $, $\beta_n $ \) are the coefficients, and \( $\epsilon $ \) is the error term.


3. **Polynomial Regression**:
   $$
   y = \beta_0 + \beta_1 x + \beta_2 x^2 + \cdots + \beta_n x^n + \epsilon
   $$
   where \( $y $\) is the dependent variable, \($ x $\) is the independent variable, \( $\beta_0, \beta_1, \beta_2, \ldots, \beta_n $\) are the coefficients, and \( $\epsilon $\) is the error term.

4. **Ridge Regression** (L2 Regularization):
   $$
   \hat{\beta} = \arg \min_{\beta} \left\{ \| y - X \beta \|^2 + \lambda \| \beta \|^2 \right\}
   $$
   where \( $y $\) is the dependent variable, \($ X $\) is the matrix of independent variables, \($ \beta $\) is the vector of coefficients, \( $\lambda $\) is the regularization parameter, and \( $\| \cdot \| $\) denotes the norm.


5. **Lasso Regression** (L1 Regularization):
   $$
   \hat{\beta} = \arg \min_{\beta} \left\{ \| y - X \beta \|^2 + \lambda \| \beta \|_1 \right\}
   $$
   where \( $y $\) is the dependent variable, \( $X $\) is the matrix of independent variables, \( $\beta $\) is the vector of coefficients, \( $\lambda $\) is the regularization parameter, and \($ \| \cdot \|_1 $\) denotes the L1 norm.


6. **Logistic Regression**:
   $$
   \text{logit}(P) = \ln \left( \frac{P}{1 - P} \right) = \beta_0 + \beta_1 x
   $$
   where \($P $\) is the probability of the dependent event occurring, \( $x $\) is the independent variable, \( $\beta_0 $\) is the intercept, \( $\beta_1 $\) is the coefficient, and \($ \text{logit}(P) $\) is the log-odds of \( $P $\).

7. **Poisson Regression**:
   $$
   \log(\lambda) = \beta_0 + \beta_1 x
   $$
   where \( $\lambda $\) is the expected count or rate of events, \( $x $\) is the independent variable, \( $\beta_0 $\) is the intercept, \( $\beta_1 $\) is the coefficient, and \( $\log(\lambda) $\) is the log of the rate.

In all of the examples above, the parameters are:

1) The $\beta_0$ the intercept in mathematical language AKA the bias in data science language.

2) The $\beta_1, \beta_2, \beta_3 $ etc are the coefficients or slopes in mathematical language AKA as the weights in Data Science language.

3) The error term $ ϵ $ is calculated as the difference between the actual label $y$ and the predicted label $\hat{y} $ which is part of what the algorithm uses to calculate the MSE(using the formular provided previously), whose value would then be used to determine how well fitted the model is to the dataset it is being trained with.
